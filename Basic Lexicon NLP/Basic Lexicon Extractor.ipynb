```python
import pandas as pd
import re
from collections import Counter
```

```python
df = pd.read_csv('twitter_training.csv',  header=None, names=['id', 'topic', 'label', 'text'])
display(df.head(5))
```

```python
def preprocess(text):
    
    if pd.isna(text):  # check if text is NaN
        return [] 
    text = re.sub(r'[^a-z\s]','', text.lower())
    return text.split()
```

```python
print(preprocess("Hello I a'm good"))
```

```python
df.dtypes
```

```python
pos_words = Counter()
neg_words = Counter()
```

```python
for _, row in df.iterrows():
    words = preprocess(row["text"])
    if row["label"] == "Positive":
        pos_words.update(words)
    elif row["label"] == "Negative":
        neg_words.update(words)
```

```python
positive_set = set(pos_words)
negative_set = set(neg_words)

neutral_words = positive_set.intersection(negative_set)
pure_positive = positive_set - neutral_words
pure_negative = negative_set - neutral_words

print("âœ… Positive words:", pure_positive)
print("âŒ Negative words:", pure_negative)
print("ðŸ˜ Neutral words:", neutral_words)
```

```python
word_sentiment = {}

for w in pure_positive:
    word_sentiment[w] = "positive"
for w in pure_negative:
    word_sentiment[w] = "negative"
for w in neutral_words:
    word_sentiment[w] = "neutral"

print(word_sentiment)
```

```python
def classify_sentence(text, word_sentiment):
    words = preprocess(text)
    score = 0
    for w in words:
        if w in word_sentiment:
            if word_sentiment[w] == "positive":
                score += 1
            elif word_sentiment[w] == "negative":
                score -= 1
    if score > 0:
        return "Positive ðŸ˜Š"
    elif score < 0:
        return "Negative ðŸ˜ "
    else:
        return "Neutral ðŸ˜"
```

```python
examples = [
    "I love borderlands and it is amazing",
    "I hate this game, it is terrible",
    "Borderlands is okay, not bad"
]

for text in examples:
    print(text, "->", classify_sentence(text, word_sentiment))
```

```python
print(classify_sentence('I love this game so much it is very beautiful and pretty', word_sentiment))
```
